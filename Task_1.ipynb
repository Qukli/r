Даны значения зарплат из выборки выпускников: 100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150. Посчитать (желательно без использования статистических методов наподобие std, var, mean) среднее арифметическое, среднее квадратичное отклонение, смещенную и несмещенную оценки дисперсий для данной выборки.

```python
зарплаты = [100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150]
# Вычисление среднего арифметического
среднее_арифметическое = sum(зарплаты) / len(зарплаты)
```

```python
# Вычисление среднего квадратичного отклонения
сумма_квадратов_разностей = sum((x - среднее_арифметическое) ** 2 for x in зарплаты)
среднее_квадратичное_отклонение = (сумма_квадратов_разностей / len(зарплаты)) ** 0.5
```

```python
# Вычисление смещенной оценки дисперсии
смещенная_дисперсия = сумма_квадратов_разностей / len(зарплаты)
```

```python
# Вычисление несмещенной оценки дисперсии
несмещенная_дисперсия = сумма_квадратов_разностей / (len(зарплаты) - 1)
```

```python
print("Среднее арифметическое:", среднее_арифметическое)
print("Среднее квадратичное отклонение:", среднее_квадратичное_отклонение)
print("Смещенная оценка дисперсии:", смещенная_дисперсия)
print("Несмещенная оценка дисперсии:", несмещенная_дисперсия)
```
